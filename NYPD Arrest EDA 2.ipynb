```python
import pandas as pd
import numpy as np 
import matplotlib.pylab as plt
import seaborn as sns
```

```python
arrest_1 = pd.read_csv(r"C:\Users\maria\Videos\NYPD_Arrest_Data__Year_to_Date_.csv")
arrest_2 = pd.read_csv(r"C:\Users\maria\Videos\NYPD_Arrests_Data__Historic_.csv")
arrest_3 = pd.read_csv(r"C:\Users\maria\Videos\NYPD_Arrests_Data__Historic_ 2.csv")
```

```python
arrest = pd.concat([arrest_1,arrest_2, arrest_3], ignore_index=True)
arrest.head().T
```

```python
arrest.replace(to_replace= "(null)",value=np.nan)
arrest
```

```python
arrest_copy = arrest[['ARREST_KEY', 'ARREST_DATE', 'OFNS_DESC', 'LAW_CAT_CD','ARREST_BORO', 'ARREST_PRECINCT', 'AGE_GROUP', 'PERP_SEX', 'PERP_RACE', 'Latitude', 'Longitude']]
arrest_copy
```

```python
for col in arrest_copy.columns:
    print(f"Column:{col:35} | type{str(arrest_copy[col].dtype):7} | missing values {arrest_copy[col].isna().sum():3d}")
```

```python
arrest_copy.dropna(subset = {'ARREST_DATE', 'ARREST_BORO','ARREST_PRECINCT','LAW_CAT_CD', 'OFNS_DESC', 'AGE_GROUP', 'PERP_RACE', 'PERP_SEX','Latitude', 'Longitude'}, inplace=True)
```

```python
arrest_copy['ARREST_BORO'].value_counts()
```

```python
Boro_key = {'B':'Bronx', 'S':'Staten Island', 'K':'Brooklyn', 'M':'Manhattan', 'Q':'Queens'}
Law_cat_key = {'F':'Felony', 'M':'Misdemeanor', 'V':'Violation', 'I':'Infraction', '9':'Unspecified'}

arrest_copy['ARREST_BORO'] = arrest_copy['ARREST_BORO'].apply(lambda x:Boro_key[x])
arrest_copy['LAW_CAT_CD'] = arrest_copy['LAW_CAT_CD'].apply(lambda x:Law_cat_key[x])
arrest_copy
```

```python
gender_key = {'M':'Male', 'F':'Female', 'U':'Unspecified'}
arrest_copy['PERP_SEX'] = arrest_copy['PERP_SEX'].apply(lambda x:gender_key[x])
```

```python
arrest_copy['ARREST_KEY'] = arrest_copy['ARREST_KEY'].astype(str)
arrest_copy['ARREST_PRECINCT'] = arrest_copy['ARREST_PRECINCT'].apply(lambda x: str(x))
```

```python
arrest_copy['ARREST_DATE'] = pd.to_datetime(arrest_copy['ARREST_DATE'])
arrest_copy['Year'] = arrest_copy['ARREST_DATE'].dt.year
arrest_copy['Month'] = arrest_copy['ARREST_DATE'].dt.month_name(locale = 'english')
arrest_copy['Day_name'] = arrest_copy['ARREST_DATE'].dt.day_name()
```

```python
arrest_copy = arrest_copy[arrest_copy["AGE_GROUP"].str.contains("<18|25-44|45-64|18-24|65+") == True]
```

```python
Year = arrest_copy.groupby('Year')['ARREST_KEY'].size().reset_index()
Year.rename(columns ={'ARREST_KEY':'No of Arrests'}, inplace=True)
Year
```

```python
fig, ax = plt.subplots(figsize = (14,6))
arrest_copy.set_index('ARREST_DATE').resample('M').size().plot(label = 'Total Arrests made Per Month', ax = ax)
arrest_copy.set_index('ARREST_DATE').resample('M').size().rolling(window = 12).mean().plot(color = 'darkblue', linewidth = 4, label = '12 Month rolling', ax = ax)
ax.set_title("Arrest Per Month", fontsize = 14, fontweight = 'bold')
ax.set(ylabel = "Total Count\n")
ax.legend(bbox_to_anchor =(1.1,1.1), frameon = False)
sns.despine(ax = ax, top = True, right = True, left = True, bottom = False)
```

```python
plt.figure(figsize = (10,5))
sns.catplot(x = 'Year', y = 'No of Arrests', data = Year, kind='bar',ax = ax, palette='brg')
plt.title('Crimes Reported in New York over the years')
plt.xlabel('No of Arrests')
plt.ylabel('Year')
plt.xticks(rotation = 45)
plt.show()
```

```python
def color_negative_red(value):
  """
  Colors elements in a dateframe
  green if positive and red if
  negative. Does not color NaN
  values.
  """

  if value <= 1000:
    color = 'green'
  elif value > 10000:
    color = 'blue'
  elif value <=10000:
    color = 'red'
  else:
    color = 'black'

  return 'color: %s' % color
```
